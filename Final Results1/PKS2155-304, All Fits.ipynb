```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mlp
from matplotlib.ticker import MultipleLocator, AutoLocator
mlp.style.use("classic")
%matplotlib inline
```

```python
#from other plots where cut out fitting curves, used https://apps.automeris.io/wpd/
#reading data from CSV files, each variable is for corresponding fit.

lmfit = pd.read_excel("/home/mher/Downloads/drive-download-20181015T123639Z-001/LMFIT Dataset.csv.xlsx", sep = ",", header = None, names = ["X", "Y"])
curve_fit = pd.read_csv("/home/mher/Downloads/drive-download-20181015T123639Z-001/Curve_fit, fixed Dataset.csv", sep = ",", header = None, names = ["X", "Y"])
article = pd.read_excel("/home/mher/Downloads/drive-download-20181015T123639Z-001/ArticleDataset.csv.xlsx", sep = ",", header = None, names = ["X", "Y"])

#ploting graphs from dataframe object

ax = lmfit.plot(x = "X",y = "Y",color = 'r',label = "LMFIT", figsize=(10,5), lw = 2.5,ls = ":")
curve_fit.plot(x = "X",y = "Y",color = 'b',label = "Curve Fit, times fixed", figsize=(10,5), lw = 2.5, ax=ax,ls= "-.")
article.plot(x = "X",y = "Y",color = 'g',label = "From Article", figsize=(10,5),  lw = 2.5, ax=ax,ls="-")

# configuring y axis
ax.set_ylim(0,4)
ytext = r'$ \mathbf {I\ \left(>200\ GeV\ [10^{-9}\ cm^{-2}\ s^{-1} ]\ \right)} $'
ax.set_ylabel(ytext, fontweight = "bold", fontsize = 15)

#configuring  x axis

xtext = "Time - MJD53944.0 [min]"
ax.set_xlabel(xtext, fontsize = 15, fontweight = "bold")
ax.set_xlim(0,140)

#configuring ticks
xminorlocator = MultipleLocator(1)
ax.xaxis.set_minor_locator(xminorlocator)
ax.tick_params(which = "major",length = 5.7 )
ax.tick_params(which = "minor", length = 3.9)


plt.savefig("/home/mher/Desktop/PKS2155-304, All Fits.pdf")
plt.show()
```
